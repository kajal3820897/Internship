```python
# Question - 1 and 2 :
import requests
from bs4 import BeautifulSoup

def search_amazon(product):
    url = f"https://www.amazon.in/s?k={product.replace(' ', '+')}"
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        results = soup.find_all('div', class_='s-result-item')
        
        for result in results:
            try:
                title = result.find('span', class_='a-size-medium').text.strip()
                price = result.find('span', class_='a-price-whole').text.strip()
                print(f"Title: {title}\nPrice: {price}\n")
            except AttributeError:
                continue
    else:
        print("Failed to retrieve Amazon search results.")


product = input("Enter the product you want to search for on Amazon.in: ")
search_amazon(product)
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd

def scrape_product_details(product):
    all_products = []
    driver = webdriver.Chrome()
    for page_number in range(1, 4):  
        url = f"https://www.amazon.in/s?k={product.replace(' ', '+')}&page={page_number}"
        driver.get(url)
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        results = soup.find_all('div', class_='s-result-item')
        for result in results:
            product_info = {}
            try:
                title = result.find('span', class_='a-size-base-plus a-color-base a-text-normal').text.strip()
                price = result.find('span', class_='a-price-whole').text.strip()[:60]
                exp_delivery = result.find('span', class_='a-color-base a-text-bold').text.strip()[:60]
                url = result.find('a', class_='a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal')['href']
                product_info['Name of the Product'] = title
                product_info['Price'] = price
                product_info['Expected_delivery'] = exp_delivery
                product_info['Url'] = url
                all_products.append(product_info)
            except AttributeError:
                continue
    driver.quit()
    return all_products


product = input("Enter the product you want to search for on Amazon.in: ")


products_details = scrape_product_details(product)


df = pd.DataFrame(products_details)


df.to_csv(f'{product}_details.csv', index=False)

print("Scraping complete. Data saved to CSV file.")
```

```python
# Question - 5 :
import re
driver_3 = webdriver.Chrome()
driver_3.get("https://www.google.co.in/maps")
time.sleep(3)

city = input('Enter City Name: ')
search_box = driver_3.find_element(By.ID, 'searchboxinput')
search_box.clear()
time.sleep(2)
search_box.send_keys(city)

button = driver_3.find_element(By.ID, 'searchbox-searchbutton')
button.click()
time.sleep(3)

try:
    
    url_string = driver_3.current_url
    print("URL Extracted:", url_string)
    
    
    lat_long = re.findall(r'@(-?\d+\.\d+),(-?\d+\.\d+)', url_string)
    if lat_long:
        lat, long = lat_long[0]
        print("Latitude = {}, Longitude = {}".format(lat, long))
    else:
        print("Latitude and longitude not found in URL.")
except Exception as e:
    print("Error:", str(e))

driver_3.quit()
```
