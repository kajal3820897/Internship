```python
# Question 5 : 
driver = webdriver.Chrome()
driver.get("https:/www.billboard.com/")
```

```python
# Click on the "Charts" option
charts = driver.find_element(By.XPATH,"/html/body/div[3]/footer/div[4]/div/div/div[2]/div[3]/div/nav[3]/ul/li[1]/a")
charts.click()

# Click on the "Hot 100" link
hot_100_link = driver.find_element(By.XPATH,'/html/body/div[3]/main/div[2]/div[1]/div[1]/div/div/div[1]/div/div[2]/span/a')
hot_100_link.click()
# Wait for the page to load
time.sleep(3)

# Initialize lists to store the details
song_names = []
artist_names = []
last_week_ranks_week_on_board = []
peak_ranks = []
```

```python
# Song name
song_title = driver.find_elements(By.XPATH,"//li[@class='lrv-u-width-100p']//li/h3")
for i in song_title:
    song_name = i.text
    song_names.append(song_name)
```

```python
# Artist name
try:
    for i in driver.find_elements(By.XPATH,"//li[@class='lrv-u-width-100p']//li//span[@class='c-label  a-no-trucate a-font-primary-s lrv-u-font-size-14@mobile-max u-line-height-normal@mobile-max u-letter-spacing-0021 lrv-u-display-block a-truncate-ellipsis-2line u-max-width-330 u-max-width-230@tablet-only']"):
        artist_names.append(i.text)
except NoSuchElementException:
        artist_names.append("--")
```

```python
# Last week rank and week on chart
last_week_rank_week = driver.find_elements(By.XPATH,"//div[@class='o-chart-results-list-row-container']//li[@class='o-chart-results-list__item // a-chart-color u-width-72 u-width-55@mobile-max u-width-55@tablet-only lrv-u-flex lrv-u-flex-shrink-0 lrv-u-align-items-center lrv-u-justify-content-center lrv-u-border-b-1 u-border-b-0@mobile-max lrv-u-border-color-grey-light u-background-color-white-064@mobile-max u-hidden@mobile-max']/span[1]")
for i in last_week_rank_week:
    last_week_board = i.text
    last_week_ranks_week_on_board.append(last_week_board)
```

```python
 # Peak rank
peak_rank = driver.find_elements(By.XPATH,"//div[@class='o-chart-results-list-row-container']//li[@class='o-chart-results-list__item // a-chart-bg-color a-chart-color u-width-72 u-width-55@mobile-max u-width-55@tablet-only lrv-u-flex lrv-u-flex-shrink-0 lrv-u-align-items-center lrv-u-justify-content-center lrv-u-background-color-grey-lightest lrv-u-border-b-1 u-border-b-0@mobile-max lrv-u-border-color-grey-light u-hidden@mobile-max']/span[1]")
for i in peak_rank:
    peak = i.text
    peak_ranks.append(peak)
```

```python
 # Print the scraped details
for i in range(len(song_names)):
        print("Song:", song_names[i])
        print("Artist:", artist_names[i])
        print("Last Week Rank  Weeks on Board:", last_week_ranks_week_on_board[i])
        print("Peak Rank:", peak_ranks[i])
        print("\n")

# Close the browser
driver.quit()
```
